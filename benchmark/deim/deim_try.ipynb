```python
# DEIM via HuggingFace — pas besoin de MMDetection
from transformers import AutoImageProcessor, AutoModelForObjectDetection
import torch
from PIL import Image
import supervision as sv

# Charger DEIM-T depuis HuggingFace
print("⏳ Chargement DEIM...")
processor = AutoImageProcessor.from_pretrained("IDEA-Research/deim_hgnetv2_t_coco")
model_deim = AutoModelForObjectDetection.from_pretrained("IDEA-Research/deim_hgnetv2_t_coco")
model_deim.eval()
print("✅ DEIM chargé !")

# Poids du modèle
import os
cache_dir = os.path.expanduser("~/.cache/huggingface/hub")
print(f"📦 Modèle chargé en mémoire")
param_size = sum(p.numel() * p.element_size() for p in model_deim.parameters()) / 1024 / 1024
print(f"📦 Taille estimée : {param_size:.1f} MB")
```

```python
# Chercher les modèles DEIM disponibles sur HuggingFace
from huggingface_hub import list_models

models = list(list_models(search="DEIM", limit=20))
for m in models:
    print(m.id)
```

Nous avons tenté d'intégrer DEIM mais ce modèle, très récent (2024), n'est pas encore disponible via les frameworks standards (HuggingFace, Ultralytics). Son utilisation nécessite un setup manuel complexe incompatible avec notre environnement Azure ML. Nous avons donc privilégié des modèles mieux intégrés comme YOLOv8/v11 et RT-DETR.
